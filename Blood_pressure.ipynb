```python
import scipy.io
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

def quick_fix():
    """Quick fix by converting continuous labels to binary classes"""
    
    # Load data (using your existing successful loading code)
    file_paths = [
        r'C:\Users\Hp\Documents\ACM_task-3\part_1.mat\part_1.mat',
        r'C:\Users\Hp\Documents\ACM_task-3\part_2.mat\part_2.mat',
        r'C:\Users\Hp\Documents\ACM_task-3\part_3.mat\part_3.mat'
    ]
    
    datasets = []
    for file_path in file_paths:
        try:
            mat_data = scipy.io.loadmat(file_path)
            data_key = [key for key in mat_data.keys() if not key.startswith('__')][0]
            raw_data = mat_data[data_key]
            
            if raw_data.dtype == 'object':
                flat_data = []
                for item in raw_data.flat:
                    if hasattr(item, '__iter__') and not isinstance(item, str):
                        try:
                            flat_item = np.array(item).flatten()
                            flat_data.extend(flat_item)
                        except:
                            continue
                
                data_array = np.array(flat_data, dtype=float)
                n_features = int(np.sqrt(len(data_array)))
                data_array = data_array[:n_features*n_features].reshape(-1, n_features)
            else:
                data_array = np.array(raw_data, dtype=float)
            
            datasets.append(pd.DataFrame(data_array))
            print(f"Loaded: {data_array.shape}")
            
        except Exception as e:
            print(f"Error loading {file_path}: {str(e)}")
    
    # Combine datasets
    min_cols = min(df.shape[1] for df in datasets)
    datasets = [df.iloc[:, :min_cols] for df in datasets]
    
    train_df = pd.concat([datasets[0], datasets[1]], ignore_index=True)
    val_df = datasets[2]
    
    # Split features and labels
    X_train = train_df.iloc[:, :-1].values.astype(float)
    y_train = train_df.iloc[:, -1].values.astype(float)
    X_val = val_df.iloc[:, :-1].values.astype(float)
    y_val = val_df.iloc[:, -1].values.astype(float)
    
    print(f"Original y_train range: {np.min(y_train):.6f} to {np.max(y_train):.6f}")
    print(f"Unique values in y_train: {len(np.unique(y_train))}")
    
    # Convert continuous labels to binary classes using median threshold
    median_threshold = np.median(y_train)
    y_train_binary = (y_train > median_threshold).astype(int)
    y_val_binary = (y_val > median_threshold).astype(int)
    
    print(f"Converted to binary classes using threshold {median_threshold:.6f}")
    print(f"Class 0: {np.sum(y_train_binary == 0)} samples")
    print(f"Class 1: {np.sum(y_train_binary == 1)} samples")
    
    # Train models
    models = {
        'Logistic Regression': LogisticRegression(random_state=42, max_iter=500),
        'Random Forest': RandomForestClassifier(n_estimators=50, random_state=42)
    }
    
    for name, model in models.items():
        try:
            print(f"\nTraining {name}...")
            model.fit(X_train, y_train_binary)
            y_pred = model.predict(X_val)
            accuracy = accuracy_score(y_val_binary, y_pred)
            print(f"{name} Accuracy: {accuracy:.4f}")
            
        except Exception as e:
            print(f"Error with {name}: {str(e)}")

if __name__ == "__main__":
    quick_fix()
```
